```python
#Ancestro en común
def ancestro(lista):
    i=0
    while(i < len(lista[0][1])):
        j=0
        ancestro = lista[j][1][i]
        while(j < len(lista)):
            if(ancestro == lista[j][1][i]):
                j = j + 1
            else:
                return lista[j][1][i-1]
        i = i + 1
    return ancestro
ances= ancestro(lista)
print("Ancestro en común = ",ances)
```
